```python
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

```

2016 Data

```python
#files to load for 2016
LA_CO_2016 = "CO-CarbonMonoxide/2016-LA-CO.csv"
LA_NO_2016 = "NitrogenDioxide/2016-LA-NO2.csv"
LA_O_2016 = "Ozone/2016-LA-Ozone.csv"
LA_PB_2016 = "PB-Lead/2016-LA-PB.csv"
LA_PM_2016 = "PM-ParticulateMatter/2016-LA-PM10.csv"
LA_SO_2016 = "SulfurDioxide/2016-LA-SO2.csv"

#Read files into Pandas dataframe

df_2016_CO = pd.read_csv(LA_CO_2016)
df_2016_NO = pd.read_csv(LA_NO_2016)
df_2016_O = pd.read_csv(LA_O_2016)
df_2016_PB = pd.read_csv(LA_PB_2016)
df_2016_PM = pd.read_csv(LA_PM_2016)
df_2016_SO = pd.read_csv(LA_SO_2016)


df_2016_PM=df_2016_PM[['Site Name','Date','Daily Mean PM10 Concentration','UNITS','DAILY_AQI_VALUE', 'SITE_LATITUDE','SITE_LONGITUDE' ]]

df_2016_CO=df_2016_CO[['Site Name','Date','Daily Max 8-hour CO Concentration','UNITS','DAILY_AQI_VALUE']]

df_2016_NO=df_2016_NO[['Site Name','Date','Daily Max 1-hour NO2 Concentration','UNITS','DAILY_AQI_VALUE']]

df_2016_O=df_2016_O[['Site Name','Date','Daily Max 8-hour Ozone Concentration','UNITS','DAILY_AQI_VALUE']]

df_2016_SO=df_2016_SO[['Site Name','Date','Daily Max 1-hour SO2 Concentration','UNITS','DAILY_AQI_VALUE']]

df_2016_PB=df_2016_PB[['Site Name','Date','Daily Mean Pb Concentration','UNITS','DAILY_AQI_VALUE']]


df_2016_PM
```

```python
left = df_2016_PM
right = df_2016_NO
LA_2016_All = pd.merge(left, right, on=['Site Name','Date'],how='outer', indicator=True) 
```

```python
left = LA_2016_All
right = df_2016_O
LA_2016_All = pd.merge(left, right, on=['Site Name','Date'],how='outer') 
```

```python
left = LA_2016_All
right = df_2016_SO
LA_2016_All = pd.merge(left, right, on=['Site Name','Date'],how='outer') 
```

```python
left = LA_2016_All
right = df_2016_CO
LA_2016_All = pd.merge(left, right, on=['Site Name','Date'],how='outer') 
```

```python
LA_2016_All.columns
```

```python
LA_2016_All.count()
```

```python
#print(ElPaso_2016_All)

```

```python
LA_2016_All.columns=['Site Name','Date','DMean PM10','PM10 Units','PM10 AQI','Lat','Lng','DMax NO2','NO2 Units','NO2 AQI','Merged','DMax Ozone','Ozone Units','Ozone AQI','DMax SO2','SO2 Units','SO2 AQI','DMax CO','CO Units','CO AQI']

```

```python
LA_2016_All.count()
```

```python
LA_2016_All.drop('Merged',axis=1,inplace=True)
LA_2016_All.to_csv("LA2016DF.csv")
```
