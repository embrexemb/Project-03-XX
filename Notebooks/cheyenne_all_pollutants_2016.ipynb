```python
#BW
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
from sklearn import datasets
import numpy as np
```

### Import files and create dataframe

```python
#files to load for 2016 Cheyenne
Cheyenne_CO_2016 = "CO-CarbonMonoxide/2016-Cheyenne-CO.csv"
Cheyenne_NO_2016 = "NitrogenDioxide/2016-Cheyenne-NO2.csv"
Cheyenne_O_2016 = "Ozone/2016-Cheyenne-Ozone.csv"
#no lead for Cheyenne
Cheyenne_PM_2016 = "PM-ParticulateMatter/2016-Cheyenne-PM10.csv"
Cheyenne_SO_2016 = "SulfurDioxide/2016-Cheyenne-SO2.csv"

#Read files into Pandas dataframe
CH_df_2016_CO = pd.read_csv(Cheyenne_CO_2016)
CH_df_2016_NO = pd.read_csv(Cheyenne_NO_2016)
CH_df_2016_O = pd.read_csv(Cheyenne_O_2016)
CH_df_2016_PM = pd.read_csv(Cheyenne_PM_2016)
CH_df_2016_SO = pd.read_csv(Cheyenne_SO_2016)
CH_df_2016_PM.head()
```

### Adds Pollutant Columns and select columns

```python
#adding Substance column into Cheyenne_CO_2016 dataframe
CH_df_2016_CO['POLLUTANT'] = 'CO'
CH_df_2016_NO['POLLUTANT'] = 'NO'
CH_df_2016_O['POLLUTANT'] = 'O'
CH_df_2016_PM['POLLUTANT'] = 'PM' 
CH_df_2016_SO['POLLUTANT'] = 'SO'
```

```python
# print Cheyenne columns
CH_df_2016_PM=CH_df_2016_PM[['Date','Site ID','Site Name','DAILY_AQI_VALUE', 'SITE_LATITUDE','SITE_LONGITUDE','COUNTY', 'STATE_CODE','STATE','POLLUTANT']]
CH_df_2016_CO=CH_df_2016_CO[['Date','Site ID','Site Name','DAILY_AQI_VALUE', 'SITE_LATITUDE','SITE_LONGITUDE','COUNTY', 'STATE_CODE','STATE','POLLUTANT']]
CH_df_2016_NO=CH_df_2016_NO[['Date','Site ID','Site Name','DAILY_AQI_VALUE', 'SITE_LATITUDE','SITE_LONGITUDE','COUNTY', 'STATE_CODE','STATE','POLLUTANT']]
CH_df_2016_O=CH_df_2016_O[['Date','Site ID','Site Name','DAILY_AQI_VALUE', 'SITE_LATITUDE','SITE_LONGITUDE','COUNTY', 'STATE_CODE','STATE','POLLUTANT']]
CH_df_2016_SO=CH_df_2016_SO[['Date','Site ID','Site Name','DAILY_AQI_VALUE', 'SITE_LATITUDE','SITE_LONGITUDE','COUNTY', 'STATE_CODE','STATE','POLLUTANT']]
CH_df_2016_PM.columns
```

### Concatenate all dataframes

```python
frames = [CH_df_2016_CO, CH_df_2016_NO, CH_df_2016_O,CH_df_2016_PM, CH_df_2016_SO]
Cheyenne_2016_All = pd.concat(frames)
Cheyenne_2016_All
```
